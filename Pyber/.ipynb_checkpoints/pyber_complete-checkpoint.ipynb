```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
full_data_set = pd.merge(city_data,ride_data, on='city')
# Display the data table for preview
full_data_set.head()
```

```python
urbanfilter = full_data_set[full_data_set["type"] == "Urban"].groupby(full_data_set["city"])
suburbanfilter = full_data_set[full_data_set["type"] == "Suburban"].groupby(full_data_set["city"])
ruralfilter = full_data_set[full_data_set["type"] == "Rural"].groupby(full_data_set["city"])
urbanfilter.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urbanx = urbanfilter['ride_id'].count()
suburbanx = suburbanfilter['ride_id'].count()
ruralx = ruralfilter['ride_id'].count()

urbany = urbanfilter['fare'].mean()
suburbany = suburbanfilter['fare'].mean()
ruraly = ruralfilter['fare'].mean()

urbanz = urbanfilter['driver_count'].mean()
suburbanz = suburbanfilter['driver_count'].mean()
ruralz = ruralfilter['driver_count'].mean()

# Build the scatter plots for each city types
plt.scatter(urbanx, urbany, marker='o', facecolors='gold', edgecolors='black', s=urbanz*10, label="Urban", alpha=.5)
plt.scatter(suburbanx, suburbany, marker='o', facecolors='lightskyblue', edgecolors='black', s=suburbanz*10, label="Suburban", alpha=.5)
plt.scatter(ruralx, ruraly, marker='o', facecolors='lightcoral', edgecolors='black', s=ruralz*10, label="Rural", alpha=.5)


# Incorporate the other graph properties
plt.title("Ride Share Data for 2018")
plt.xlabel("Total Rides per City")
plt.ylabel("Average Fare in Dollars")
plt.grid()

# Create a legend
legend = plt.legend(title="City Type", loc='best')
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]

# Incorporate a text label regarding circle size
plt.text(0,12, "*Circle Size is relative to the number of drivers per city.*", color='red', size=12)

# Save Figure
plt.savefig("../Images/2018RideShareBubbleChart", bbox_inches='tight')
plt.show()
```

## Total Fares by City Type

```python
# (to get percents i need totals for all, and each type)
total_all_fares = full_data_set["fare"].sum()
urban_fares_total = full_data_set.loc[full_data_set['type'] == 'Urban','fare'].sum()
suburban_fares_total = full_data_set.loc[full_data_set['type'] == 'Suburban','fare'].sum()
rural_fares_total = full_data_set.loc[full_data_set['type'] == 'Rural','fare'].sum()

# Calculate Type Percents 
urban_fares_pct = (urban_fares_total/total_all_fares)*100
suburban_fares_pct = (suburban_fares_total/total_all_fares)*100
rural_fares_pct = (rural_fares_total/total_all_fares)*100


# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
sizes = [urban_fares_pct, suburban_fares_pct, rural_fares_pct]
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=120)
plt.title("Percentage of Fares by City Type")
plt.axis('equal')


# Save Figure
plt.savefig("../Images/2018pct_fares_bycitytype", bbox_inches='tight')

plt.show()


```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_all_rides = full_data_set["ride_id"].count()
urban_rides = full_data_set.loc[full_data_set['type'] == 'Urban','ride_id'].count()
suburban_rides = full_data_set.loc[full_data_set['type'] == 'Suburban','ride_id'].count()
rural_rides = full_data_set.loc[full_data_set['type'] == 'Rural','ride_id'].count()

urban_rides_pct = (urban_rides/total_all_rides)*100
suburban_rides_pct = (suburban_rides/total_all_rides)*100
rural_fares_pct = (rural_rides/total_all_rides)*100

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
sizes = [urban_rides, suburban_rides, rural_rides]
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=120)
plt.title("Percentage of Rides by City Type")
plt.axis('equal')

# Save Figure
plt.savefig("../Images/2018pct_rides_bycitytype", bbox_inches='tight')


plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urban_drivers = urbanfilter["driver_count"].mean()
urban_drivers_total = urban_drivers.sum()
suburban_drivers = suburbanfilter["driver_count"].mean()
suburban_drivers_total = suburban_drivers.sum()
rural_drivers = ruralfilter["driver_count"].mean()
rural_drivers_total = rural_drivers.sum()
total_all_drivers = urban_drivers_total + suburban_drivers_total + rural_drivers_total



urban_drivers_pct = (urban_drivers_total/total_all_drivers)*100
suburban_drivers_pct = (suburban_drivers_total/total_all_drivers)*100
rural_drivers_pct = (rural_drivers_total/total_all_drivers)*100



# # Build Pie Charts
labels = ["Urban", "Suburban", "Rural"]
sizes = [urban_drivers_total, suburban_drivers_total, rural_drivers_total]
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=120)
plt.title("Percentage of Drivers by City Type")
plt.axis('equal')
# Save Figure
plt.savefig("../Images/2018pct_drivers_bycitytype", bbox_inches='tight')
plt.show()
```
